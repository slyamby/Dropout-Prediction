### Student's Dropout - Tensorflow

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

df = pd.read_csv('students_dropout.csv')
df
```

```python
df = df.drop(['Course'],axis=1)
```

### Converting the Target Column from Categorical to Numerical

```python
df_binary = df[df['target'].isin(['Dropout', 'Graduate'])]
df_binary.loc[:, 'target'] = df['target'].map({'Dropout': 0, 'Graduate': 1})
df_binary
```

### Extracting the inputs and targets

```python
inputs = df_binary.iloc[:, :-1]
inputs
```

```python
targets = df_binary.iloc[:, -1]
targets
```

```python
targets.value_counts()
```

### Balance Dataset

```python
num_one_targets = int(np.sum(targets))

# Set a counter for targets that are 0 (meaning that the customer did not convert)
zero_targets_counter = 0

indices_to_remove = []

for i in range(targets.shape[0]):
    if targets.iloc[i] == 0: # This tells Pandas: “Get me the i-th row, regardless of index labels.
        zero_targets_counter += 1
        if zero_targets_counter > num_one_targets:
            indices_to_remove.append(i)
            
# Create two new variables, one that will contain the inputs, and one that will contain the targets.
# We delete all indices that we marked "to remove" in the loop above.
unscaled_inputs_equal_priors = np.delete(inputs, indices_to_remove, axis=0)
targets_equal_priors = np.delete(targets, indices_to_remove, axis=0)
```

### Standardize the inputs

```python
scaled_inputs = preprocessing.scale(unscaled_inputs_equal_priors)
```

### Shuffle the data

```python
shuffled_indices = np.arange(scaled_inputs.shape[0])
np.random.shuffle(shuffled_indices)

# Use the shuffled indices to shuffle the inputs and the targets
shuffled_inputs = scaled_inputs[shuffled_indices]
shuffled_targets = targets_equal_priors[shuffled_indices]
```

### Split the dataset into train, validation, test

```python
from sklearn.model_selection import train_test_split

# First split off the test set
train_temp, test_inputs, target_temp, test_targets = train_test_split(shuffled_inputs, shuffled_targets, test_size=0.2, random_state=42)

# Then split temp into train and validation
train_inputs, val_inputs, train_targets, val_targets = train_test_split(train_temp, target_temp, test_size=0.25, random_state=42)

print(train_inputs.shape[0])
print(val_inputs.shape[0])
print(test_inputs.shape[0])
```

### Save the three datasets

```python
np.savez('school_dropout_data_train', inputs=train_inputs, targets=train_targets)
np.savez('school_dropout_data_validation', inputs=val_inputs, targets=val_targets)
np.savez('school_dropout_data_test', inputs=test_inputs, targets=test_targets)
```
