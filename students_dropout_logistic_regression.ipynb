### Students Dropout - Logistic Regression

```python
import pandas as pd
```

```python
raw_csv_data = pd.read_csv('students_dropout.csv')
```

```python
raw_csv_data
```

```python
#raw_csv_data = raw_csv_data.drop(['Course'],axis=1)
df = raw_csv_data.copy()
df
```

```python
df.info()
```

```python
df.describe()
```

### Create the Targets

```python
df['target'].value_counts()
```

```python
df_binary = df[df['target'].isin(['Dropout', 'Graduate'])]
```

```python
df_binary.info()
```

```python
df_binary.loc[:, 'target_encoded'] = df_binary['target'].map({'Dropout': 0, 'Graduate': 1})
```

```python
df_binary
```

```python
df_binary = df_binary.drop(['target'],axis=1)
```

```python
data_with_targets = df_binary.copy()
```

```python
data_with_targets.info()
```

```python
targets = data_with_targets['target_encoded']
targets.shape
```

### Select the inputs for the regression

```python
inputs = data_with_targets.iloc[:, :30]
inputs
```

```python
unscaled_inputs = inputs.copy()
```

### Standardize the inputs

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = scaler.transform(unscaled_inputs)
```

```python
scaled_inputs
```

```python
scaled_inputs.shape
```

```python
targets.shape
```

### Split the data into train and test and shuffle

```python
from sklearn.model_selection import train_test_split
```

### Split

```python
x_train, x_test, y_train, y_test  = train_test_split(scaled_inputs, targets, train_size=0.8, random_state=42)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

### Logistic Regression with Sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import numpy as np
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

### Manually check the accuracy

```python
model_outputs = reg.predict(x_train) # testing the model with the training data
model_outputs
```

```python
model_outputs == y_train
```

```python
np.sum(model_outputs == y_train)
```

```python
model_outputs.shape[0]
```

```python
np.sum(model_outputs == y_train) / model_outputs.shape[0]
```

### Finding Intercepts and Coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_inputs.columns.values # using scaled inputs will give you an error because it is a numpy array
```

```python
feature_names = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns=['Feature_name'], data=feature_names)

summary_table['Coefficient'] = np.transpose(reg.coef_)

summary_table
```

```python
summary_table.index = summary_table.index+1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

### Testing the model

```python
# Assess the test accuracy of the model
reg.score(x_test, y_test)
```

```python
# find the predicted probability of each class
predicted_proba = reg.predict_proba(x_test)

predicted_proba
```

```python
predicted_proba.shape
```

```python
# Select on probabilities referring to 1
predicted_proba[:, 1]
```
